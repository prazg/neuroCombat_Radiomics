```python
# Import necessary libraries
import numpy as np
import pandas as pd
from neuroCombat import neuroCombat
```

```python
data = pd.read_csv(r"C:\Users\pg22\OneDrive - King's College London\Documents\PhD Data\Matched data\Matched_1.5TTradiomics_ex_Outliers_52_T.csv", low_memory=False)
data.head()
```

```python
# Make the first column the index
data = data.set_index(data.columns[0])

data.head()
```

```python
# Define the batch variable
batch = data[['Site']]
batch.head()
```

```python
# This assumes that the rest of the columns in your data are the radiomics features
features = data.drop(['Site'], axis=1)
features.head()
```

```python
# Specify the batch columns
batch_cols = ['Site']

# Ensure that 'batch_cols' is a subset of 'batch'
batch = batch[batch_cols]

print(batch['Site'].value_counts())
```

```python
# Transpose the features DataFrame
features = features.T
```

```python
# Perform neuroComBat
data_combat = neuroCombat(dat=features, covars=batch, batch_col=batch_cols)
```

```python
# Extract the data from the returned object
data = data_combat['data']
```

```python
# Convert the transposed data to a DataFrame
df_combat = pd.DataFrame(data.T, index=features.columns, columns=features.index)
```

```python
# Print the first few rows of the df_combat DataFrame
print(df_combat.head())
```

```python
# Print the first few rows of the original features DataFrame
print(features.head())
```
